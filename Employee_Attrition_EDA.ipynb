# Machine Learning

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,LinearRegression
from sklearn.metrics import balanced_accuracy_score,accuracy_score,mean_squared_error
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier,RandomForestClassifier,AdaBoostClassifier
import time
from sklearn.svm import SVC
```

```python
# going to use the dataset having the dummy columns
data = './IBM_dummied.csv'
df = pd.read_csv(data, index_col=0)
df=df.astype(float)
df.shape
```

```python
df.columns
```

```python
df.rename({'attrition_yes':'attrition'}, axis=1, inplace=True)
```

```python
# separating target from the dataset
X = df.drop('attrition', axis=1)
y = df.attrition
```

```python
def model_testing(model):
    #Split the data using test_train_split
    x_train,x_test,y_train,y_test = train_test_split(X, y, test_size=0.2, random_state=4)
    #Fit the model using the data which was splitted
    model.fit(x_train,y_train)
    #Print Score of the model
    score = model.score(x_test,y_test)
    return (score)
    #print(y_pred)
    
def plotGraph(keys, values, label, title):
    names = keys
    y_pos = np.arange(len(keys))
    data = values
    plt.barh(y_pos, data, align='center', alpha=0.6)
    plt.yticks(y_pos, names)
    plt.xlabel(label)
    plt.title(title)
    plt.show()
    
#get the models
model_dict = {
    "KNN": KNeighborsClassifier(n_neighbors = 3),
    "LogisticRegression": LogisticRegression(),
    "RandomForest": RandomForestClassifier(),
    "AdaBoostClassifier": AdaBoostClassifier(),
    "DecisionTreeClassifier" : DecisionTreeClassifier(),
    "GradientBoostingClassifier":GradientBoostingClassifier(),
    "SVC":SVC(kernel="rbf", C=0.025, probability=True),
    "GaussianNB":  GaussianNB()
}

preds = {}
for modelname in model_dict:    
    model = model_dict[modelname]
    score=model_testing(model)
    #preds.append(score) 
    preds[modelname]=score
    #model_stats_score[modelname] = preds
    print("average for {} : {}".format(modelname,preds[modelname]))
```

```python
log_cols=["Classifier", "Accuracy" ]
log = pd.DataFrame(columns=log_cols)
for i in model_dict:
    log_entry=pd.DataFrame([[model_dict[i],preds[i]]],columns=log_cols)
    log=log.append(log_entry)

log1 = log.sort_values(by='Accuracy',ascending=False)
```

```python
# accuracy per classifier
plt.figure(figsize=(12,4))
fig=sns.barplot(x='Accuracy', y='Classifier',data=log1 ,palette=sns.color_palette("Reds", n_colors=len(model_dict.keys())))

fig.set_xlabel('Accuracy', fontsize=15)
plt.xticks(fontsize=12)
fig.yaxis.labelpad = 30
fig.set_ylabel('Classifier Model', fontsize=15)
fig.yaxis.labelpad = 30
plt.yticks(fontsize=12)
plt.title('Classifier Accuracy',fontsize=20);
fig.yaxis.labelpad = 30

plt.show()
```
